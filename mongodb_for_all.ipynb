# Funciones CRUD básicas

```python
#Insertar un estudiante
def insertar_estudiante(nombre, edad, carrera):
    estudiante = {"nombre": nombre, "edad": edad, "carrera": carrera}
    estudiantes.insert_one(estudiante)
    print(f"Estudiante {nombre} insertado.")
```

```python
# Listar estudiantes de forma amigable
def listar_estudiantes():
    print("\n--- Listado de estudiantes: ---")
    estudiantes_db = list(estudiantes.find())
    if len(estudiantes_db) == 0:
        print("  No hay estudiantes registrados aún.")
    for est in estudiantes_db:
        # Extraemos las claves por si alguna falta para evitar errores
        nombre = est.get("nombre", "Sin nombre")
        edad = est.get("edad", "Sin edad")
        carrera = est.get("carrera", "Sin carrera")
        print(f"🔹 Nombre: {nombre} | Edad: {edad} | Carrera: {carrera}")
```

```python
# Actualizar información de un estudiante
def actualizar_estudiante(nombre_actual, campo, nuevo_valor):
    estudiantes.update_one({"nombre": nombre_actual}, {"$set": {campo: nuevo_valor}})
    print(f"La información de {nombre_actual} fue actualizada exitosamente.")
```

```python
# Eliminar un estudiante
def eliminar_estudiante(nombre):
    estudiantes.delete_one({"nombre": nombre})
    print(f"Estudiante {nombre} eliminado.")
```

```python
def menu():
    while True:
        print("\n--- MENÚ DE ESTUDIANTES ---")
        print("1. Insertar estudiante")
        print("2. Listar estudiantes")
        print("3. Actualizar información de un estudiante")
        print("4. Eliminar estudiante")
        print("5. Salir")
        
        opcion = input("Selecciona una opción (1-5): ")
        
        if opcion == '1':
            nombre = input("Ingresa el nombre del estudiante: ")
            try:
                edad = int(input(f"Ingresa la edad de {nombre}: "))
                carrera = input(f"Ingresa la carrera de {nombre}: ")
                insertar_estudiante(nombre, edad, carrera)
            except ValueError:
                print("La edad debe ser un número entero válido.")
        elif opcion == '2':
            listar_estudiantes()
        elif opcion == '3':
            nombre = input("Ingresa el nombre del estudiante que deseas actualizar: ")
            print("\n¿Qué dato deseas actualizar?")
            print("1. Nombre")
            print("2. Edad")
            print("3. Carrera")
            sub_opcion = input("Selecciona una opción (1-3): ")
            
            if sub_opcion == '1':
                nuevo_nombre = input(f"Ingresa el nuevo nombre para {nombre}: ")
                actualizar_estudiante(nombre, "nombre", nuevo_nombre)
            elif sub_opcion == '2':
                try:
                    nueva_edad = int(input(f"Ingresa la nueva edad para {nombre}: "))
                    actualizar_estudiante(nombre, "edad", nueva_edad)
                except ValueError:
                    print("La edad debe ser un número entero válido.")
            elif sub_opcion == '3':
                nueva_carrera = input(f"Ingresa la nueva carrera para {nombre}: ")
                actualizar_estudiante(nombre, "carrera", nueva_carrera)
            else:
                print("Opción no válida. Volviendo al menú principal...")
        elif opcion == '4':
            nombre = input("Ingresa el nombre del estudiante a eliminar: ")
            eliminar_estudiante(nombre)
        elif opcion == '5':
            print("Saliendo del programa...")
            break
        else:
            print("Opción no válida. Inténtalo de nuevo.")

# Ejecutar el menú
menu()
```
